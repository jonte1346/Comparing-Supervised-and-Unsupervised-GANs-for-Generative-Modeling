```python
#Step 3.A: List and Preview Your Images
import os
from PIL import Image

image_folder = '/content/drive/MyDrive/MBL 559E - Group Project/Fifth Test/U2Net_Cl_resized'
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Preview one image (optional)
img = Image.open(os.path.join(image_folder, image_files[0]))
img.show()
```

```python
#Step 4.A: Set Up the Augmentation Pipeline
import albumentations as A
import cv2

augment = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.Rotate(limit=15, p=0.5),
    A.RandomBrightnessContrast(p=0.3),
    A.GaussNoise(p=0.1),
    A.RandomScale(scale_limit=0.2, p=0.3),
])

output_folder = '/content/drive/MyDrive/MBL 559E - Group Project/Fifth Test/U2Net_Cl_resized_augm'
os.makedirs(output_folder, exist_ok=True)
```

```python
#Step 5.A: Run Augmentation and Save Results
for fname in image_files:
    img_path = os.path.join(image_folder, fname)
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    for i in range(5):  # 5 augmented versions per image
        augmented = augment(image=image)['image']
        out_fname = f"{os.path.splitext(fname)[0]}_aug{i}.jpg"
        out_path = os.path.join(output_folder, out_fname)
        # Convert RGB to BGR for saving with OpenCV
        cv2.imwrite(out_path, cv2.cvtColor(augmented, cv2.COLOR_RGB2BGR))
```
