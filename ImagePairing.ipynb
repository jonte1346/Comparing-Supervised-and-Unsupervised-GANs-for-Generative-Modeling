```python
import cv2
import os

# Input directories
masks_dir = '/content/drive/MyDrive/MBL 559E - Group Project/Fifth Test/U2Net_Cl_resized_segmn'
images_dir = '/content/drive/MyDrive/MBL 559E - Group Project/Fifth Test/U2Net_Cl_resized_augm'
output_dir = '/content/drive/MyDrive/MBL 559E - Group Project/Fifth Test/paired_images'
os.makedirs(output_dir, exist_ok=True)

for filename in os.listdir(masks_dir):
    if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
        base = os.path.splitext(filename)[0].replace('_mask', '')  # adjust if your mask names end with _mask
        # For masks, adapt extension if needed (usually .png)
        mask_path = os.path.join(masks_dir, filename)
        # Search for image with same base name (could be .png or .jpg)
        possible_exts = ['.jpg', '.jpeg', '.png']
        img_path = None
        for ext in possible_exts:
            test_path = os.path.join(images_dir, base + ext)
            if os.path.exists(test_path):
                img_path = test_path
                break
        if img_path is None:
            print(f"Image not found for {filename}")
            continue
        img = cv2.imread(img_path)
        mask = cv2.imread(mask_path)
        # Resize both to 256x256
        img = cv2.resize(img, (256, 256), interpolation=cv2.INTER_AREA)
        mask = cv2.resize(mask, (256, 256), interpolation=cv2.INTER_NEAREST)
        # Concatenate side-by-side: MASK (LEFT), FACADE (RIGHT)
        pair = cv2.hconcat([mask, img])
        out_name = base + '_pair.png'
        out_path = os.path.join(output_dir, out_name)
        cv2.imwrite(out_path, pair)
```
